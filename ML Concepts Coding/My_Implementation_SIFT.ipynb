```python
import cv2
from skimage.io import imread,imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img3=cv2.imread('Konark.png')
img4=cv2.imread('Konark3.png')
#imshow(img1)
#imshow(img2)
```

```python
img1=cv2.cvtColor(img3,cv2.COLOR_BGR2GRAY)
img2=cv2.cvtColor(img4,cv2.COLOR_BGR2GRAY)
```

```python
#imshow(img2)
```

```python
sift=cv2.xfeatures2d.SIFT_create()
```

```python
keypoints_1,descriptors_1=sift.detectAndCompute(img1,None)
keypoints_2,descriptors_2=sift.detectAndCompute(img2,None)
```

```python
bf=cv2.BFMatcher(cv2.NORM_L1,crossCheck=True)
```

```python
matches=bf.match(descriptors_1,descriptors_2)
```

```python
matches=sorted(matches,key=lambda x:x.distance)
```

```python
img3=cv2.drawMatches(img3,keypoints_1,img4,keypoints_2,matches[:20],img4,flags=2)
```

```python
plt.figure(figsize=(10,20))
plt.imshow(img3)
```
