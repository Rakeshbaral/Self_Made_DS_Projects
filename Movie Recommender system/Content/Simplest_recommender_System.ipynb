# Content Based Simple Movie Recommendation System

<b>Regarding Dataset: 100,000 ratings and 3,600 tag applications applied to 9,000 movies by 600 users

```python
import pandas as pd
import numpy as np

ratings_df=pd.read_csv("ratings.csv")
movies_df=pd.read_csv("movies.csv")
tags_df=pd.read_csv("tags.csv")
movies_df.head()
```

```python
#pd.concat(frames,keys(***These are levels),axis=0 or 1,join=inner/outer,)
#.reindex(d1.index)
#d1.append(df2)This is same as concat
#d1.append([d2,d3])
#d1.concat([d1,d2]).ignore_index=True

#Merge function is fully meant for the join operation
#df.merge(left,right,on=(if the column names are keys on which the dfs will be joined),left_on=,right_on,left_index=True(Take the left index)
#right_index=,how=inner,outer,left,right
```

```python
df_result=movies_df.merge(ratings_df,on='movieId')
df_result=df_result.merge(tags_df,on='movieId')
ratings_df.shape
```

```python
df_result
```

```python
#Creating a dataframe of the rating mean of each movie. But the demaerit is some movies is not watched by many people by 
#have been give rating which does not infer to be popular movie
rating=pd.DataFrame(df_result.groupby('title')['rating'].mean())
```

```python
#The number of ratings given by the users plays primary role here so that we acn guess if the movie has been seen by 
#most number of people
rating_count=pd.DataFrame(df_result.groupby('title')['rating'].count().sort_values(ascending=False))
```

```python
rating['number of total ratings']=pd.DataFrame(df_result.groupby('title')['rating'].count())
```

```python
rating.sort_values(by='number of total ratings',ascending=False)
```

```python
#rating['user_id']=df_result['userId']
#Creating a pivot table having individual movies and individual used in the columns and rows correspondingly.
#It will have lot of NaN values because most of the movies are not watched by most of the users
movie_matrix=df_result.pivot_table(index='userId_y',columns='title',values='rating')
movie_matrix.columns
```

```python
movie_matrix
```

```python
movie_name=input("Please put a movie name to see the recommended movies to watch next")
```

```python
movie_name='(500) Days of Summer (2009)'
```

```python
#Collecting the movie ratings in a series
movie=movie_matrix[movie_name]
```

```python
#Most of the movies will have null values in most of the rows, we need to remove the null values
movie.dropna(inplace=True)
```

```python
movie
```

```python
#Applying the correlation with the rest of the movie columns to get the similar movies
Sim_movie=movie_matrix.corrwith(movie)
```

```python
Sim_movie=pd.DataFrame(Sim_movie,columns=['Correlation'])
```

```python
#Getting similar movies with most number of ratings in descending order i.e. populat movies having highest ratings
Sim_movie=Sim_movie.join(rating['number of total ratings']).sort_values('number of total ratings',ascending=False)
```
